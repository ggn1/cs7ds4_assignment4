```python
# Correlation matrix.
df_corr = df.drop(
    ['breed', 'size', 'coat_type', 'coat_length', 'image']
    + [f'rank_{year}' for year in range(2013, 2021)]
, axis=1).corr()
plt.figure(figsize=(16, 6))
sns.heatmap(
    df_corr, vmin=-1, vmax=1, 
    annot=True, cmap='coolwarm',
    mask=np.triu(np.ones_like(df_corr))
)
```
